# Detektimi i outliers

```python
# Calculate and count outliers for each numerical feature
outliers_count_per_feature = {}

# Iterate through each numerical column
for column_name in dataset.select_dtypes(include=['number']):
    # Calculate the IQR for the current column
    Q1 = dataset[column_name].quantile(0.25)
    Q3 = dataset[column_name].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers for the current column
    outliers = dataset[(dataset[column_name] < lower_bound) | (dataset[column_name] > upper_bound)]
    
    # Count the number of outliers for the current column
    outliers_count_per_feature[column_name] = len(outliers)

     # Visualize boxplot with outliers highlighted
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=dataset[column_name], orient="v", fliersize=5, color="skyblue", linewidth=1.5)
    plt.title(f'Boxplot of {column_name} with Outliers Highlighted')
    plt.ylabel(column_name)
    
    # Highlight outliers
    sns.swarmplot(x=outliers[column_name], color="red", size=5)
    
    plt.show()

# Print the count of outliers for each numeric feature
print("Count of outliers for each numeric feature:")
for column_name, count in outliers_count_per_feature.items():
    print(f"{column_name}: {count}")


```

# Largimi i outliers

```python
# Remove outliers from the dataset
outliers_removed_dataset = dataset.copy()  # Create a copy of the original dataset

# Iterate through each numerical column
for column_name in dataset.select_dtypes(include=['number']):
    # Calculate the IQR for the current column
    Q1 = dataset[column_name].quantile(0.25)
    Q3 = dataset[column_name].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Remove outliers for the current column
    outliers_removed_dataset = outliers_removed_dataset[(outliers_removed_dataset[column_name] >= lower_bound) & 
                                                        (outliers_removed_dataset[column_name] <= upper_bound)]

# Print the shape of the dataset before and after removing outliers
print("Shape of dataset before removing outliers:", dataset.shape)
print("Shape of dataset after removing outliers:", outliers_removed_dataset.shape)
```

# Faza **II** Analiza dhe evaluimi 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

# Encode all categorical variables
label_encoder = LabelEncoder()
dataset['interior_color'] = label_encoder.fit_transform(dataset['interior_color'])
dataset['exterior_color'] = label_encoder.fit_transform(dataset['exterior_color'])
dataset['brand'] = label_encoder.fit_transform(dataset['brand'])
dataset['model'] = label_encoder.fit_transform(dataset['model'])
dataset['engine_size_category'] = label_encoder.fit_transform(dataset['engine_size_category'])
dataset['drivetrain_category'] = label_encoder.fit_transform(dataset['drivetrain_category'])
dataset['fuel_category'] = label_encoder.fit_transform(dataset['fuel_category'])

# Select features and target variable
X = dataset.drop('first_owner', axis=1)  # Features
y = dataset['first_owner']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Number of rows in the original dataset
total_rows = len(dataset)

# Number of rows for training
rows_training = len(X_train)

# Number of rows for testing
rows_testing = len(X_test)

print("Total rows:", total_rows)
print("Rows for training:", rows_training)
print("Rows for testing:", rows_testing)
```

# Aplikimi i algoritmit **Random Forest Classifier**

```python
 
# Initialize and train the Random Forest Classifier
from sklearn.metrics import f1_score, recall_score, precision_score

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

  
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Calculate F1-score
f1score = f1_score(y_test, y_pred)

# Calculate recall
recall = recall_score(y_test, y_pred)

# Calculate precision
precision = precision_score(y_test, y_pred)

print("F1-score (Random Forest):", f1score)
print("Recall (Random Forest):" , recall)
print("Precision (Random Forest):", precision)
```

# Aplikimi i algoritmit **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# Initialize and train the Logistic Regression model
logreg_model = LogisticRegression(random_state=42)
logreg_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_logreg = logreg_model.predict(X_test)

# Calculate accuracy
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print("Accuracy (Logistic Regression):", accuracy_logreg)

# Calculate F1-score
f1score_logreg = f1_score(y_test, y_pred_logreg)

# Calculate recall
recall_logreg = recall_score(y_test, y_pred_logreg)

# Calculate precision
precision_logreg = precision_score(y_test, y_pred_logreg)

print("F1-score (Logistic Regression):", f1score_logreg)
print("Recall (Logistic Regression):", recall_logreg)
print("Precision (Logistic Regression):", precision_logreg)
```

# Aplikimi i algoritmit **SVM**

```python
from sklearn.svm import SVC

# Initialize and train the SVM model
svm_model = SVC(kernel='rbf', random_state=42)  # 'rbf' kernel is commonly used
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_svm = svm_model.predict(X_test)

# Calculate accuracy
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("Accuracy (SVM):", accuracy_svm)

# Calculate F1-score
f1score_svm = f1_score(y_test, y_pred_svm)

# Calculate recall
recall_svm = recall_score(y_test, y_pred_svm)

# Calculate precision
precision_svm = precision_score(y_test, y_pred_svm)

print("F1-score (SVM):", f1score_svm)
print("Recall (SVM):", recall_svm)
print("Precision (SVM):", precision_svm)
```

# Aplikimi i algoritmit  **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB

# Initialize and train the Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred_nb = nb_model.predict(X_test)

# Calculate accuracy
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("Accuracy (Naive Bayes):", accuracy_nb)

# Calculate F1-score
f1score_nb = f1_score(y_test, y_pred_nb)

# Calculate recall
recall_nb = recall_score(y_test, y_pred_nb)

# Calculate precision
precision_nb = precision_score(y_test, y_pred_nb)

print("F1-score (Naive Bayes):", f1score_nb)
print("Recall (Naive Bayes):", recall_nb)
print("Precision (Naive Bayes):", precision_nb)
```

# Paraqitja e linjave të grupuara për të vizualizuar performancën e secilit algoritëm në metrika të ndryshme.

```python
import matplotlib.pyplot as plt

metrics = ['Accuracy', 'F1-score', 'Precision', 'Recall']
random_forest_values = [accuracy, f1score, precision, recall]
logistic_regression_values = [accuracy_logreg, f1score_logreg, precision_logreg, recall_logreg]
svm_values = [accuracy_svm, f1score_svm, precision_svm, recall_svm]
naive_bayes_values = [accuracy_nb, f1score_nb, precision_nb, recall_nb]

# Set the positions for the bars
positions = range(len(metrics))

# Plotting
plt.figure(figsize=(12, 8))

plt.bar([p - 0.2 for p in positions], random_forest_values, width=0.2, label='Random Forest')
plt.bar([p for p in positions], logistic_regression_values, width=0.2, label='Logistic Regression')
plt.bar([p + 0.2 for p in positions], svm_values, width=0.2, label='SVM')
plt.bar([p + 0.4 for p in positions], naive_bayes_values, width=0.2, label='Naive Bayes')

plt.xticks(positions, metrics)
plt.title('Performance Comparison of Algorithms')
plt.ylabel('Value')
plt.legend()
plt.show()
```
