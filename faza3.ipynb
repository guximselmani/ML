# Detektimi i outliers

```python
# Calculate and count outliers for each numerical feature
outliers_count_per_feature = {}

# Iterate through each numerical column
for column_name in dataset.select_dtypes(include=['number']):
    # Calculate the IQR for the current column
    Q1 = dataset[column_name].quantile(0.25)
    Q3 = dataset[column_name].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define the lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers for the current column
    outliers = dataset[(dataset[column_name] < lower_bound) | (dataset[column_name] > upper_bound)]
    
    # Count the number of outliers for the current column
    outliers_count_per_feature[column_name] = len(outliers)

     # Visualize boxplot with outliers highlighted
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=dataset[column_name], orient="v", fliersize=5, color="skyblue", linewidth=1.5)
    plt.title(f'Boxplot of {column_name} with Outliers Highlighted')
    plt.ylabel(column_name)
    
    # Highlight outliers
    sns.swarmplot(x=outliers[column_name], color="red", size=5)
    
    plt.show()

# Print the count of outliers for each numeric feature
print("Count of outliers for each numeric feature:")
for column_name, count in outliers_count_per_feature.items():
    print(f"{column_name}: {count}")


```

```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

dataset.drop(columns=['engine_size_category', 'usb_port','stability_control','bluetooth'  ,'turbo','fuel_category'], inplace=True)
X = dataset.drop('first_owner', axis=1)  # Features
y = dataset['first_owner']  # Target variable
 
# Define a function to train and evaluate models
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1score = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    return accuracy, f1score, recall, precision

# Function to train and evaluate models with different test sizes
def train_and_evaluate_models(X, y, test_sizes, models):
    # Scale features
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=max(test_sizes.values()), random_state=42)
    
    # Apply SMOTE to balance the classes in the training data
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    
    # Train and evaluate models for each test size
    for name, model in models.items():
        test_size = test_sizes[name]
        print(f"Results for {name} with test size = {test_size}:")
        X_train_temp, _, y_train_temp, _ = train_test_split(X_train_resampled, y_train_resampled, test_size=test_size, random_state=42)
        accuracy, f1score, recall, precision = train_and_evaluate_model(model, X_train_temp, X_test, y_train_temp, y_test)
        print(f"  Accuracy: {accuracy}")
        print(f"  F1-score: {f1score}")
        print(f"  Recall: {recall}")
        print(f"  Precision: {precision}")
        print()

# Initialize models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(random_state=42),
    "SVM": SVC(kernel='rbf', random_state=42),
    "Naive Bayes": GaussianNB()
}

# Specify different test sizes for each algorithm
test_sizes = {
    "Random Forest": 0.2,
    "Logistic Regression": 0.3,
    "SVM": 0.2,
    "Naive Bayes": 0.3
}

# Train and evaluate models with different test sizes
train_and_evaluate_models(X, y, test_sizes, models)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_curve, auc
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import MinMaxScaler

# Define a function to train and evaluate models
def train_and_evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1score = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    
    # Calculate ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(X_test)[:,1])
    roc_auc = auc(fpr, tpr)
    
    return accuracy, f1score, recall, precision, fpr, tpr, roc_auc

# Function to train and evaluate models with different test sizes
def train_and_evaluate_models(X, y, test_sizes, models):
    # Scale features
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Initialize dictionaries to store evaluation metrics for each algorithm
    metrics_dict = {metric: {name: [] for name in models} for metric in ['Accuracy', 'F1-score', 'Precision', 'Recall', 'ROC AUC']}
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=max(test_sizes.values()), random_state=42)
    
    # Apply SMOTE to balance the classes in the training data
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    
    # Train and evaluate models for each test size
    for name, model in models.items():
        test_size = test_sizes[name]
        X_train_temp, _, y_train_temp, _ = train_test_split(X_train_resampled, y_train_resampled, test_size=test_size, random_state=42)
        
        # Train the model
        accuracy, f1score, recall, precision, fpr, tpr, roc_auc = train_and_evaluate_model(model, X_train_temp, X_test, y_train_temp, y_test)
        
        # Store evaluation metrics in the dictionary
        metrics_dict['Accuracy'][name] = accuracy
        metrics_dict['F1-score'][name] = f1score
        metrics_dict['Recall'][name] = recall
        metrics_dict['Precision'][name] = precision
        metrics_dict['ROC AUC'][name] = roc_auc
        
        # Plot ROC curve
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random Guess')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.grid()
    plt.show()

    # Plotting other metrics
    plt.figure(figsize=(12, 8))

    # Define the metrics
    metrics = ['Accuracy', 'F1-score', 'Precision', 'Recall', 'ROC AUC']
    colors = ['blue', 'orange', 'green', 'red', 'purple']

    # Plot each metric
    for i, metric in enumerate(metrics):
        plt.subplot(2, 3, i+1)
        plt.bar(models.keys(), [metrics_dict[metric][name] for name in models], color=colors[i])
        plt.title(metric)
        plt.xlabel('Algorithm')
        plt.ylabel('Value')

    plt.tight_layout()
    plt.show()

# Initialize models
models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Logistic Regression": LogisticRegression(random_state=42),
    "SVM": SVC(kernel='rbf', random_state=42, probability=True),  # Set probability=True for SVC to enable predict_proba
    "Naive Bayes": GaussianNB()
}

# Specify different test sizes for each algorithm
test_sizes = {
    "Random Forest": 0.4,
    "Logistic Regression": 0.4,
    "SVM": 0.4,
    "Naive Bayes": 0.4
}

# Train and evaluate models with different test sizes
train_and_evaluate_models(X, y, test_sizes, models)
```
